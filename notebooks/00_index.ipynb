# An Introduction to exatomic

Exatomic is a package that facilitates data analysis for computational chemistry.

Key features of exatomic include:
- Parsing of quantum chemistry software output
- Visualization of molecules, orbitals
- Dataframe based data organization

## [Basics of exatomic](./01_basics.ipynb)

- Tutorial of basic usage and syntax
- Demonstrates how to load/manipulate dataframe data

## [XYZ Tutorial](./02_xyz.ipynb)

- Load an XYZ trajectory
- Visualize the structure
- Examine basic data objects associated with/derived from the structure

## [Visualization of Molecular Orbitals](./03_orbitals.ipynb)

- Tutorial on parsing data from wave function theory software
- Visualization of molecular orbitals
- Computation of overlap, density matrix, and other manipulations stemming from the basis set, order, and coefficient matrix
